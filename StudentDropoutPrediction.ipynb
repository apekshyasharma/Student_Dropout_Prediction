## **Step 1: Exploratory Data Analysis**
Exploratory Data Analysis (EDA) is an approach to analyzing data sets to summarize their main characteristics, often with visual methods. A primary goal of EDA is to see what the data can tell us beyond the formal modeling or hypothesis testing task. It helps in:

*   **Understanding the data's structure and content:** Identifying data types, missing values, and potential errors.
*   **Discovering patterns and relationships:** Finding correlations, trends, and anomalies.
*   **Formulating hypotheses:** Generating ideas for further investigation or modeling.
*   **Preparing data for modeling:** Informing decisions about data cleaning, transformation, and feature engineering.

Common techniques include:

*   **Summary statistics:** Mean, median, mode, standard deviation, quartiles.
*   **Data visualization:** Histograms, scatter plots, box plots, bar charts, heatmaps.

```python

# Create derived features from important columns
df_clean['Performance_Decline'] = (
    df_clean['Curricular units 1st sem (grade)'] -
    df_clean['Curricular units 2nd sem (grade)']
)

df_clean['Age_Group'] = pd.cut(df_clean['Age at enrollment'],
                                bins=[0, 22, 30, 100],
                                labels=['Traditional', 'Young_Adult', 'Mature'])

df_clean['Academic_Efficiency'] = (
    df_clean['Curricular units 2nd sem (approved)'] /
    df_clean['Curricular units 2nd sem (evaluations)'].replace(0, 1)
)

df_clean['Financial_Risk'] = (
    (~df_clean['Tuition fees up to date'].astype(bool)).astype(int) +
    df_clean['Debtor'].astype(int) +
    (~df_clean['Scholarship holder'].astype(bool)).astype(int)
)
```
